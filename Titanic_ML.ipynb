```python
#Q8. Drop the Column
df.drop('Cabin', axis=1, inplace=True)
df.isnull().sum()
```

```python
#Q9. Handle Missing Values on df['Embarked']
df['Embarked'].mode()
df['Embarked'].fillna('S', inplace=True)
df.isnull().sum()
```

```python
#Q10. Handle Missing Values on df['Age']
df['Age'].fillna(df['Age'].mean(), inplace=True)
df.isnull().sum()
```
